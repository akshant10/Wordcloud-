<b>what are world clouds? </b><p>
Word clouds (also known as text clouds or tag clouds) work in a simple way: the more a specific word appears in a source of textual data (such as a speech, blog post, or database), the bigger and bolder it appears in the word cloud.

```python
import pandas as pd

#you can download this csv from github or you can import your own dataset
#I'm using dataset about reviews of note20 ultra from flipkart
mydata1 = pd.read_csv('C:/Users/aksha/Downloads/file1.csv')

mydata1.head() 
```

```python
#import following libraries

from wordcloud import WordCloud, ImageColorGenerator
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
```

```python
all_reviews=[] #get all reviews in a list
for i in range(0,len(mydata1['reviews'])):
    all_reviews.append(mydata1['reviews'][i])
```

```python
wc = WordCloud(background_color='black') #create and initialize a WordCloud object
wc.generate(all_reviews[0]) #pass the text
plt.figure(figsize=(20, 8))
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.show()
```
